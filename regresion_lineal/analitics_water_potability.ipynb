2. Limpieza de Datos: Datos faltantes, datos err√≥neos y datos nulos


Trabajamos con los datos de Trihalomethanes [Se empieza por esta por que es la que menos datos nulos tiene]


```python
data['Trihalomethanes'].describe()
```

```python
# Generamos un DataFrame con los datos de Trihalomethanes que no sean nulos
data_not_null = data.drop(data[data.Trihalomethanes.isnull()].index)
data_not_null.head(10)

# Generamos un DataFrame con los datos de Trihalomethanes que sean nulos
data_null = data.drop(data_not_null.index)

# Creamos las estructuras conteniendo las caracter√≠sticas y la variable objetivo y entrenamos el modelo, para crear el feature set [ft] este se debe de generar con las columnas que no tienen datos nulos para reducir
# los datos cesgados

ft = ['Hardness', 'Solids', 'Chloramines', 'Conductivity',
      'Organic_carbon', 'Turbidity', 'Potability']
y = data_not_null['Trihalomethanes']
x = data_not_null[ft]

# Usamos el modelo de regresi√≥n lineal para predecir el valor de Trihalomethanes con la librer√≠a sklearn
model = LinearRegression()
model.fit(x, y)

# Armamos un DataFrame con los datos de Trihalomethanes que sean nulos
dataFrameAux = data_null[ft]
predicted_Trihalomethanes = model.predict(dataFrameAux)

# Una vez obtenido el valor de Trihalomethanes, se agrega a la columna de Trihalomethanes del DataFrame original
data.loc[data.Trihalomethanes.isnull(
), 'Trihalomethanes'] = predicted_Trihalomethanes

# Validamos que ahora Trihalomethanes no tenga valores nulos
data.isnull().sum()
```

```python
# Verificamos que ahora Trihalomethanes no tenga valores negativos
data.loc[data['Trihalomethanes'] < 0]
```

Trabajamos con los datos de ph


```python
# Validamos cada una de las columnas que presentan datos nulos, esto nos ayudara a normalizar los datos

data['ph'].describe()
```

```python
# Generamos un DataFrame con los datos de ph que no sean nulos
data_not_null = data.drop(data[data.ph.isnull()].index)
# Generamos un DataFrame con los datos de ph que sean nulos
data_null = data.drop(data_not_null.index)

# Creamos las estructuras conteniendo las caracter√≠sticas y la variable objetivo y entrenamos el modelo, para crear el feature set [ft] este se debe de generar con las columnas que no tienen datos nulos para reducir
# los datos cesgados

ft = ['Trihalomethanes', 'Hardness', 'Solids', 'Chloramines', 'Conductivity',
      'Organic_carbon', 'Turbidity', 'Potability']
y = data_not_null['ph']
x = data_not_null[ft]

# Usamos el modelo de regresi√≥n lineal para predecir el valor de ph con la librer√≠a sklearn
model = LinearRegression()
model.fit(x, y)

# Armamos un DataFrame con los datos de ph que sean nulos
dataFrameAux = data_null[ft]
predicted_ph = model.predict(dataFrameAux)

# Una vez obtenido el valor de ph, se agrega a la columna de ph del DataFrame original
data.loc[data.ph.isnull(), 'ph'] = predicted_ph

# Validamos que ahora ph no tenga valores nulos
data.isnull().sum()
```

```python
# Verificamos que ahora ph no tenga valores negativos
data.loc[data['ph'] < 0]
```

```python
# Verificamos las muestras erroneas para ph, estas que tienen un valor > 11
data.loc[data['ph'] > 11]

# Se observa que la cantidad de muestras erroneas no es muy grande por lo tanto se puede corregir cambiando estos valores por la media de los datos para ph
```

```python
# Cambiamos las muestras erroneas por la media de los datos para ph
mean_ph = data['ph'].mean()
data.loc[data['ph'] > 11] = mean_ph
```

```python
# Verificamos las muestras erroneas para ph, estas que tienen un valor > 11
data.loc[data['ph'] > 11]
```

Ahora trabajamos con la columna Sulfate, que es la que mas datos tiene nulos , primero se trabajaron con ph y Trihalomethanes para reducir la cantidad de error que nos pueda dar la regresion lineal


```python
data['Sulfate'].describe()
```

```python
# Generamos un DataFrame con los datos de Sulfate que no sean nulos
data_not_null = data.drop(data[data.Sulfate.isnull()].index)
data_not_null.head(10)

# Generamos un DataFrame con los datos de Sulfate que sean nulos
data_null = data.drop(data_not_null.index)

# Creamos las estructuras conteniendo las caracter√≠sticas y la variable objetivo y entrenamos el modelo, para crear el feature set [ft] este se debe de generar con las columnas que no tienen datos nulos para reducir
# los datos cesgados

ft = ['Trihalomethanes', 'ph', 'Hardness', 'Solids', 'Chloramines', 'Conductivity',
      'Organic_carbon', 'Turbidity', 'Potability']
y = data_not_null['Sulfate']
x = data_not_null[ft]

# Usamos el modelo de regresi√≥n lineal para predecir el valor de Sulfate con la librer√≠a sklearn
model = LinearRegression()
model.fit(x, y)

# Armamos un DataFrame con los datos de Sulfate que sean nulos
dataFrameAux = data_null[ft]
predicted_Sulfate = model.predict(dataFrameAux)

# Una vez obtenido el valor de Sulfate, se agrega a la columna de Sulfate del DataFrame original
data.loc[data.Sulfate.isnull(), 'Sulfate'] = predicted_Sulfate

# Validamos que ahora Sulfate no tenga valores nulos
data.isnull().sum()
```

```python
# Verificamos que ahora Sulfate no tenga valores negativos
data.loc[data['Sulfate'] < 0]
```

3. Realizar el an√°lisis estad√≠stico descriptivo a por lo menos tres variables cuantitativas del dataset


```python
# Se realiza nuevamente un analisis general, con los datos corregidos
data.describe()
```

```python
# Grafica de Tendencia

fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(9, 9))
axes = axes.flatten()

col_num = data.select_dtypes(include=['float64', 'int64']).columns
col_num = col_num.drop(['Conductivity'])

for i, col in enumerate(col_num):

    axes[i].plot(data[col])
    axes[i].set_title(col, fontsize=7, fontweight="bold")
    axes[i].tick_params(labelsize=6)



fig.tight_layout()
fig.suptitle('Grafica de tendencia',
             fontsize=10, fontweight="bold")
```

```python
# Grafica de Histograma

fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(9, 5))
axes = axes.flatten()

col_num = data.select_dtypes(include=['float64', 'int64']).columns
col_num = col_num.drop(['Conductivity'])

for i, col in enumerate(col_num):
    sns.histplot(
        data=data,
        x=col,
        stat="frequency",
        kde=True,
        color=(list(plt.rcParams['axes.prop_cycle'])*2)[i]["color"],
        line_kws={'linewidth': 2},
        alpha=0.3,
        ax=axes[i]
    )
    axes[i].set_title(col, fontsize=7, fontweight="bold")
    axes[i].tick_params(labelsize=6)
    axes[i].set_xlabel("")

fig.tight_layout()
plt.subplots_adjust(top=0.9)
fig.suptitle('Frecuencia con la que se presenta cada variable',
             fontsize=10, fontweight="bold")
```
